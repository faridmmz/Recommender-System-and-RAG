```python
def chunk_data(products, max_tokens_per_chunk):
    """
    Splits a list of products into smaller chunks.
    Each chunk has a total token count less than max_tokens_per_chunk.
    """
    chunks = []
    current_chunk = []
    current_token_count = 0

    for product in products:
        # Create a single string from the product dictionary values
        product_string = ' '.join(str(value) for value in product.values())
        product_token_count = len(product_string.split())  # Estimate token count

        if current_token_count + product_token_count > max_tokens_per_chunk:
            chunks.append(current_chunk)
            current_chunk = []
            current_token_count = 0

        current_chunk.append(product)
        current_token_count += product_token_count

    if current_chunk:
        chunks.append(current_chunk)

    return chunks

def create_prompt_from_chunk(chunk, user_query):
    """
    Converts a chunk of product data into a formatted prompt string.
    """
    # Concatenate product details into a single string
    product_details = []
    for product in chunk:
        product_info = f"Product Name: {product['Product Name']}, Product Price: {product['Product Price']}, Product Rating: {product['Product Rating']}, Product Description: {product['Product Description']}"
        product_details.append(product_info)

    system_prompt = ' '.join(product_details)

    # Create the prompt template
    prompt_template = f'''SYSTEM: You are a helpful, respectful and honest assistant and expert of game related products. You know that {system_prompt}.

USER: {user_query}

ASSISTANT:
'''
    return prompt_template

chunks = chunk_data(products, max_tokens_per_chunk=2000)  # Split the products into chunks
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np

# Initialize Sentence Transformers model
model = SentenceTransformer('all-MiniLM-L6-v2')

def generate_embeddings(chunk):
    # Combine the product descriptions into a list
    descriptions = [product['Product Description'] for product in chunk]
    # Generate embeddings for the descriptions
    embeddings = model.encode(descriptions, convert_to_tensor=True)
    # Convert embeddings to a list of numpy arrays
    embeddings_np = [embedding.cpu().numpy() for embedding in embeddings]
    return embeddings_np

# Generate embeddings for each chunk
chunk_embeddings = [generate_embeddings(chunk) for chunk in chunks]
```

```python
index_config = {
    "settings": {
        "number_of_shards": 1,
        "number_of_replicas": 0
    },
    "mappings": {
        "properties": {
            "embedding": {
                "type": "dense_vector",
                "dims": 384
            },
            "product_info": {
                "type": "text"
            }
        }
    }
}

# Create an index for embeddings
index_name = "product_embeddi"
es.indices.create(index=index_name, body=index_config, ignore=400)
```

```python
def save_embedding_to_elasticsearch(embedding, product_info, index_name="product_embeddi"):
    # Prepare the document
    doc = {
        "embedding": embedding.tolist(),
        "product_info": product_info
    }
    # Index the document in Elasticsearch
    es.index(index=index_name, body=doc)

# Iterate over chunks and their embeddings
for chunk, embeddings in zip(chunks, chunk_embeddings):
    for product, embedding in zip(chunk, embeddings):
        product_info = f"Product Name: {product['Product Name']}, Product Description: {product['Product Description']}"
        save_embedding_to_elasticsearch(embedding, product_info)
```

```python
def search_similar_products(query, model, es, index_name="product_embeddi", top_k=3):
    # Generate an embedding for the query
    query_embedding = model.encode(query, convert_to_tensor=False)

    # Elasticsearch query to find the most similar products
    script_query = {
        "script_score": {
            "query": {"match_all": {}},
            "script": {
                "source": "cosineSimilarity(params.query_vector, 'embedding') + 1.0",
                "params": {"query_vector": query_embedding}
            }
        }
    }

    response = es.search(
        index=index_name,
        body={
            "size": top_k,
            "query": script_query,
            "_source": {"includes": ["product_info"]}
        }
    )

    # Extract product information from the response
    product_infos = [hit["_source"]["product_info"] for hit in response['hits']['hits']]
    return product_infos
```

```python
user_query = "Game Console suggestion"
relevant_products = search_similar_products(user_query, model, es)

# Create a prompt for Llama model
system_prompt = ' '.join(relevant_products)
prompt_template = f'''SYSTEM: You are a helpful, respectful and honest assistant and expert of game related products. You know that {system_prompt}.

USER: {user_query}

ASSISTANT:
'''
```
